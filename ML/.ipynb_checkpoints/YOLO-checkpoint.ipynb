```python
from PIL import Image
import numpy as np
import pandas as pd
import random
from os import listdir
%matplotlib inline
import matplotlib.pyplot as plt

from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.models import Model, Sequential
from keras.layers import *
import keras.backend as K
import keras.optimizers as Optimizers
import keras.callbacks
```

## YOLO

```python
grid_w = 11
grid_h = 3

def load_photos(directory):
    images = []
    x_labels_loc = []
    y_labels_loc = []
    class_labels_loc = []
    dir_list = listdir(directory)
    random.shuffle(dir_list)
    for name in dir_list:
        filename = directory + '/' + name
        # load an image from file
        image = load_img(filename)
        # convert the image pixels to a numpy array
        image = img_to_array(image)
        images.append(image)
        # get labels
        x_labels_loc.append(int(float(name.split('_')[1]) * grid_w))
        y_labels_loc.append(int(float(name.split('_')[2].split('.png')[0]) * grid_h))
    return images, x_labels_loc, y_labels_loc
```

```python
dataset = "Yolo_Datas/"

# load images from both train and test groups
directory = dataset + 'Train'
images, x_labels_loc, y_labels_loc = load_photos(directory)
nb_images = len(images)
print('Loaded Images and labels for training: %d' % nb_images)
```

```python
#Normalise images
images = np.array(images)
images = images
images /= 255.0

#convert datas to dummyvalues
x_labels_loc = np.array(pd.get_dummies(x_labels_loc))
y_labels_loc = np.array(pd.get_dummies(y_labels_loc))
```
